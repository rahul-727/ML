```python
bin_edges = [0,3,5,np.inf]
bin_labels = ['Low Happiness', 'Medium Happiness', 'High Happiness']
```

```python
df_copy = df.copy()
df_copy['Happiness_Score'] = pd.cut(df['Score'], bins=bin_edges, labels=bin_labels)
```

```python
df_copy
```

```python
encoder = OneHotEncoder()
encoder.fit(df_copy[['Country or region']])
encoded_data = encoder.transform(df_copy[['Country or region']])
```

```python
df_copy
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df_copy.columns:
  le=LabelEncoder()
  df_copy[col]=le.fit_transform(df_copy[col])
```

```python
from scipy.stats import zscore
zscore_df = df_copy.apply(zscore)
```

```python
o_c=((zscore_df>3)|(zscore_df<-3)).sum()
o_c
```

```python
df_copy.columns
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df_copy['Country or region'], df_copy['Overall rank'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-Square Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
data = df_copy[['GDP per capita', 'Social support','Healthy life expectancy']]
cov_matrix = np.cov(data, rowvar=False)

cov_matrix
```

```python
#A heatmap of the covariance matrix.
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=data.columns, yticklabels=data.columns)

plt.show()
```

```python
#A bar chart showing the count of countries in each Happiness Score group.

happiness_score_groups = df_copy["Happiness_Score"].unique()
country_counts = df_copy["Happiness_Score"].value_counts()

plt.bar(happiness_score_groups, country_counts)
plt.xlabel("Happiness Score Group")
plt.ylabel("Number of Countries")
plt.show()
```
