```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Clothing E-Commerce Reviews.csv')
df
```

```python
df = df.drop(columns = ['Title','Review Text'])
```

```python
age_bin_edges = [18,25, 35, 45, np.inf]
age_bin_labels = ['Young', 'Middle-Aged-Adults', 'older-adults','senior-citizens']
```

```python
df_copy = df.copy()
df_copy['Age_Category'] = pd.cut(df['Age'], bins=age_bin_edges, labels=age_bin_labels)
```

```python
df_copy
```

```python
age_group_counts = df_copy['Age_Category'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(age_bin_labels, age_group_counts , color='skyblue')
plt.xlabel('Age Groups')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customers Across Age Groups')
plt.show()
```

```python
df_copy['repeatedpurchase'] = df['Rating']
df_copy
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df_copy.columns:
  le=LabelEncoder()
  df[col]=le.fit_transform(df_copy[col])
```

```python
age_rating_data = df_copy[['Age', 'Rating']]
cov_matrix = np.cov(age_rating_data, rowvar=False)

cov_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=age_rating_data.columns, yticklabels=age_rating_data.columns)

plt.title('Age vs Rating)')
plt.show()
```

```python
df_copy['Gender'] = ['Male' if age < 30 else 'Female' for age in df['Age']]
```

```python
df_copy
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df_copy['Gender'], df_copy['repeatedpurchase'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
contingency_table.plot(kind='bar', stacked=True)
plt.title("Gender vs. Repeated Purchase")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Repeated Purchase")
plt.show()
```

```python
replacement_value = 0
df_copy['repeatedpurchase'] = df['repeatedpurchase'].replace(2, replacement_value)
```

```python
df_copy
```

```python
from scipy.stats import ttest_ind

repeated_purchases = df_copy[df_copy['repeatedpurchase'] == 1]['Rating']
not_repeated_purchase = df_copy[df_copy['repeatedpurchase']==0]['Rating']
t_statistic, p_value = ttest_ind(repeated_purchases, not_repeated_purchase)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data = [repeated_purchases, not_repeated_purchase], palette='Set3')
plt.title('Product Ratings for Repeated vs Not Repeated Purchases')
plt.xticks([0, 1], ['Repeated Purchases', 'Not Repeated Purchases'])
plt.ylabel('Product Rating')
plt.show()
```
